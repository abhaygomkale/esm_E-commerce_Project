# 🎯 E-Commerce Customer Behavior Analysis – Project Story

---

## 🟣 Introduction (Problem Statement)

Sir, the main problem in e-commerce is that companies generate a large amount of data, but they often lack clear visibility into customer behavior.  
Because of this, they miss opportunities in improving customer retention and increasing revenue.

---

## 🟣 Objective

The objective of this project is to analyze customer transaction data, identify purchasing patterns, and generate insights that support data-driven business decisions.

---

## 🟣 Dataset Overview

The dataset contains around **50,000 transaction records** with features such as:

- Customer_ID  
- Customer_Name  
- Quantity  
- Country  
- Product_Category  
- Purchase_Amount  
- Payment_Method  
- Transaction_Date  

Since it is real-world data, it contains missing values and inconsistencies.

---

## 🟣 Data Cleaning

To prepare the dataset for analysis:

- Handled missing values  
- Standardized column formats  
- Converted data types (especially date columns)  
- Removed invalid or null entries  

After cleaning, the dataset was reduced to approximately **38,000 valid records**.

---

## 🟣 Exploratory Data Analysis (EDA)

EDA was performed to understand patterns and trends:

- Revenue by product category  
- Sales across different countries  
- Payment method usage  
- Monthly sales trends  

---

## 🟣 Key Findings

- The **Toys category** generates the highest revenue (~3.75M)  
- **Mexico** is the top-performing country in terms of sales  
- **UPI** is the most commonly used payment method  
- Sales trend is mostly stable with a noticeable **spike in late 2024**

---

## 🟣 Business Interpretation

The spike in late 2024 can be explained by the global holiday season, especially **Christmas and New Year**, when customers tend to purchase more, particularly products like toys for gifting.

Additionally:

- E-commerce platforms run major **discount campaigns**  
- Increased **marketing and promotions** boost sales  

These insights help businesses:

- Focus on high-performing categories (Toys)  
- Target strong markets (Mexico)  
- Plan campaigns during peak seasons  

---

## 🟣 Machine Learning Approach

To extend the analysis:

- Applied **One-Hot Encoding** for categorical variables  
- Performed **Feature Engineering** (Year, Month, Day extraction)  
- Used **Train-Test Split (80-20)**  
- Applied **Feature Scaling (StandardScaler)**  

Model Used:
- **Linear Regression** to predict Purchase Amount  

---

## 🟣 Model Performance

- MAE: 242.84  
- RMSE: 284.45  
- R² Score: -0.0017  

### Interpretation:
The model shows very poor performance, indicating a weak relationship between features and purchase amount.

---

## 🟣 Key Learning

- Linear Regression underfits this dataset  
- Customer behavior is complex and non-linear  
- Real-world data requires advanced modeling  

---

## 🟣 Future Improvements

- Use advanced models (Random Forest, XGBoost)  
- Apply customer segmentation (Clustering)  
- Improve feature engineering with behavioral features  

---

This project successfully:

- Analyzed customer behavior

- Generated meaningful business insights  

It highlights the importance of data cleaning, analysis, and proper model selection in real-world data science problems.
